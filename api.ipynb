1. Use the API to get all available data for the GDP per capita, PPP (constant 2017 international $) indicator. Hint: this indicator has code "NY.GDP.PCAP.PP.KD". Adjust the query parameters so that you can retrieve all available rows. Convert the results to a DataFrame.

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import json
import time
```

```python
endpoint = 'https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?format=json&per_page=17556'
response = requests.get(endpoint)
```

```python
res = response.json()[1]
```

```python
gdp_df = pd.DataFrame(res)
```

```python
gdp_df.drop(columns=['indicator'], inplace=True)
```

```python
pd.json_normalize(gdp_df['country'])
```

```python
gdp_df = pd.merge(pd.json_normalize(gdp_df['country']).rename(columns={'value': 'country'}), gdp_df.drop(columns=['country', 'unit', 'obs_status', 'decimal']), left_index=True, right_index=True)
```

```python
gdp_df.rename(columns={'value':'gdp_per_capita'}, inplace=True)
```

```python
gdp_df
```

5. If you haven't already done so and you would like to get some additional practice using loops, use the page parameter in order to pull all records. Do not change the value of the per_page parameter. You will likely need to utilize a loop of some kind in order to pull all records.

```python
endpoint = 'https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?format=json'
response = requests.get(endpoint)
```

```python
pages = response.json()[0]['pages']
```

```python
data = response.json()[1]
```

```python
# data
```

```python
for i in range(2, pages+1):
    while True:
        try:
            endpoint = f'https://api.worldbank.org/v2/country/all/indicator/NY.GDP.PCAP.PP.KD?format=json&page={i}'
            response = requests.get(endpoint)
            data += response.json()[1]
            break
        except json.JSONDecodeError:
            continue
```

```python
data_df = pd.DataFrame(data)
```
